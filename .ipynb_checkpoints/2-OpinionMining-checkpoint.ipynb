```python
import re #regex library
```

```python
def GenerateAspectsFile(inputFile,tagsToInclude, OutputFile):
    pat = re.compile('\"([^\"]*)\"') # This regex pattern is to extract lemma values from the output file generated by opinion finder tool.
    sent_transaction_aspects=[]
    with open(OutputFile, 'w') as aspectsFile:
        for line in open(inputFile):
            li=line.strip()
            if not li.endswith("GATE_Sentence"):
                slc_col=pat.findall(line.rstrip()) # slc_col: string, lemma ,category collection
                if len(slc_col)==3 and slc_col[2] in tagsToInclude and slc_col[1] and slc_col[1] !='t': #slc_col[0] means string value slc_col[1] means lemma value, slc_col[2] means category value
                    sent_transaction_aspects.append(slc_col[1])
            elif len(sent_transaction_aspects) > 0:
                aspectsFile.write(','.join(sent_transaction_aspects))
                aspectsFile.write('\n');
                sent_transaction_aspects=[]
        aspectsFile.write(','.join(sent_transaction_aspects)) # Finally write the last transaction aspects too!
        print('output file generate at ' +OutputFile)
```

```python
noun_tags=['NN','NNS','NNP','NNPS']
GenerateAspectsFile('Output\\gate_default',noun_tags, 'Output\\aspects.csv')
```
