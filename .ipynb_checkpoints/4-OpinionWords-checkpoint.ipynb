Exercise 4 [15 points] Use the identified frequent aspects in Exercise 3 to find candidate opinion words.
These are the adjectives or adverbs which are near the frequent aspects identified in Exercise 3 (within
five words on both sides of a frequent aspect). (Submit the list of top-10 most frequently used opinion
words for both 0.5% and 1% support.)

Note: If the same opinion word is near to one or more frequent aspect in same sentence, then you should
count it only once.

```python
import nltk
from nltk.corpus import stopwords
from ast import literal_eval
import collections
import re
from nltk import sent_tokenize
from nltk import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
def GenerateDistinctOpinionsPerTransaction(inputFile,tagsToInclude, OutputFile):
    opinions=set()
    pat = re.compile('\"([^\"]*)\"') # This regex pattern is to extract lemma values from the output file generated by opinion finder tool.
    sent_transaction_opinions=set() #If the same opinion word is near to one or more frequent aspect in same sentence, then you should count it only once. Hence using Sets
    with open(OutputFile, 'w') as opinionFile:
        for line in open(inputFile):
            li=line.strip()
            if not li.endswith("GATE_Sentence"):
                slc_col=pat.findall(line.rstrip()) # slc_col: string, lemma ,category collection
                if len(slc_col)==3 and slc_col[2] in tagsToInclude and slc_col[1]: #slc_col[0] means string value slc_col[1] means lemma value, slc_col[2] means category value
                    sent_transaction_opinions.add(slc_col[1])
                    opinions.add(slc_col[1])
            elif len(sent_transaction_opinions) > 0:
                opinionFile.write(','.join(sent_transaction_opinions))
                opinionFile.write('\n');
                sent_transaction_opinions=set()
        opinionFile.write(','.join(sent_transaction_opinions)) # Finally write the last transaction aspects too!
        for opinion in sent_transaction_opinions:
            opinions.add(opinion)
        print('output file generate at ' + OutputFile)
        return opinions
```

```python
# nltk.help.upenn_tagset()
# or https://cs.nyu.edu/grishman/jet/guide/PennPOS.html
ad_verb_jective_tags=['JJ','JJR','JJS','RB','RBR','RBS']
opinions=GenerateDistinctOpinionsPerTransaction('Output\\gate_default',ad_verb_jective_tags, 'Output\\4_Adverb_Adjectives.csv')
```

```python
with open('data\\CanonG3_untagged.txt','r') as unTag:
    sentences=sent_tokenize(unTag.read())
    print(len(sentences))
```

```python
def GetTopNFreqOpinionCollection(inputFile,N):
    freq_aspects_nouns = []
    lemma = WordNetLemmatizer()
    freq_opinions=[]
    #freq_opinions=set()
    with open('Output\\'+inputFile,'r') as inpfile:
        for line in inpfile.readlines():
            freq_aspects_nouns.append(re.findall(r"['\"](.*?)['\"]", line)[0])
    #for every sentence tokenize to words and lemmatize them
    for sentence in sentences:
        words= list(map(lemma.lemmatize,word_tokenize(sentence)))
        #for this sentence related words, find the adjectives +/- words from the words if it is a noun
        #first get all nouns in the sentence of words i.e, instead of pos tagging we can get common list of words from words collection and freq_aspects_nouns collection
        #and get the frequent opinions from the same
        #Note: If the same opinion word is near to one or more frequent aspect in same sentence, then you should count it only once. So use set.
        sent_nouns_coll=list(set(words).intersection(freq_aspects_nouns))
        if(len(sent_nouns_coll)!=0):
            for freq_opinion in list(set(words).intersection(opinions)):
                freq_opinions.append(freq_opinion)
                #freq_opinions.add(freq_opinion)
    #print all freq opinions to file
    counter=collections.Counter(list(freq_opinions))
    allFreqOpinions='Output\\4_All_Frequent_Opinions_'+inputFile
    with open(allFreqOpinions,'w') as allFreqOp:
        allFreqOp.write(str(counter).replace('Counter({','').replace(',','\n').replace('})','').replace(" '","'"))
    #print top 10 frequent opinions to file
    with open('Output\\4_Top_10_Frequent_Opinions_'+inputFile, 'w') as topNFreqOpinion:
        topNFreqOpinion.write(str(counter.most_common(N)).replace('(','').replace('),','\n').replace('[','').replace(')]','').replace(" '","'"))
    print('Output Generated in file: '+allFreqOpinions)
    return freq_opinions
```

```python
#freq_opinions
freq_opinions_0_5_percent=GetTopNFreqOpinionCollection('3_All_Frequent_Aspects3_ARM_FP_GROWTH_0_5_percent_support.txt',10)
collections.Counter(list(freq_opinions_0_5_percent)).most_common(10)
```

```python
#freq_opinions
freq_opinions_1_percent=GetTopNFreqOpinionCollection('3_All_Frequent_Aspects3ARM_FP_GROWTH_1_percent_support.txt',10)
collections.Counter(list(freq_opinions_1_percent)).most_common(10)
```
